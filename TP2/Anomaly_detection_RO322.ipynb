```python
import networkx as nx
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_to_graph = ""
G = nx.read_edgelist(path_to_graph+'net.txt') 
cores = nx.algorithms.core.core_number(G)
print('Number of nodes :',G.number_of_nodes())
print('Number of edges :', G.number_of_edges())
print("core value",max(cores.values()))
```

```python
df = pd.DataFrame({'Degree': dict(G.degree).values(),'Coreness': list(cores.values())})
```

```python
fig, ax = plt.subplots(figsize=(15,7.5))
fig = sns.regplot(ax=ax, x="Degree", y="Coreness", data=df, fit_reg=False)
# fig = sns.scatterplot(ax=ax, x="Degree", y="Coreness", data=df, hue =df.index, palette='Spectral')
fig.set(xscale="log")
fig.set(yscale="log")
fig.set(title='Detection des anomalies')
```
